```python
# Libraries
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(-30, 40, 710)
```

```python
y = 3 * (np.sin(2 * x))
```

```python
fig, axarr = plt.subplots(3, 2, figsize = (12, 6))
axarr[0, 0].plot(x, 3*np.sin(2*x))
axarr[0, 0].set_title("3*sin(2x)")
axarr[0, 1].plot(x, np.tan(x**3))
axarr[0, 1].set_title("3*tan(x^3)")
axarr[1, 0].plot(x, np.cos(x))
axarr[1, 0].set_title("Cos")
axarr[1, 1].plot(x, np.pi/x**-3)
axarr[1, 1].set_title("pi/x^-3")
plt.tight_layout()
```
